## Final Analysis

1) Urban areas have the most riders, possibly because they also have the most drivers available. It could be interesting to cross reference this data with car ownership per household and see if that has a correlation with Pyber ridership.

2) Fares are highest in rural areas. This could be due to rural areas being less dense than urban areas and therefore rides being longer or due to the lack of drivers driving prices up (surge pricing?).

3) Rural areas have only 2.6% of the drivers, meaning that the higher rural fares could be caused by scarcity and the lower ridership could be due to wait times.

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Split by city type
urban_type = merged.loc[merged['type']=='Urban']
suburban_type = merged.loc[merged['type']=='Suburban']
rural_type = merged.loc[merged['type']=='Rural']


urban_cities = urban_type.groupby('city')
suburban_cities = suburban_type.groupby('city')
rural_cities = rural_type.groupby('city')


# Find x coordinates (Number of Rides per city)
# Find y coordinates (Average Fare)
# Find size (Drivers by city)


for city in urban_cities:
    urban_city_rides = urban_type.groupby('city')['ride_id'].count()
    urban_city_price = urban_type.groupby('city')['fare'].mean()
    urban_city_drivers = urban_type.groupby('city')['driver_count'].mean()

for city in suburban_cities:
    suburban_city_rides = suburban_type.groupby('city')['ride_id'].count()
    suburban_city_price = suburban_type.groupby('city')['fare'].mean()
    suburban_city_drivers = suburban_type.groupby('city')['driver_count'].mean()

for city in rural_cities:
    rural_city_rides = rural_type.groupby('city')['ride_id'].count()
    rural_city_price = rural_type.groupby('city')['fare'].mean()
    rural_city_drivers = rural_type.groupby('city')['driver_count'].mean()


# Build the scatter plots for each city types
urban = plt.scatter(urban_city_rides, urban_city_price, c='coral', s=urban_city_drivers*7, edgecolors='black', label='Urban', alpha=0.75)
suburban = plt.scatter(suburban_city_rides, suburban_city_price, c='skyblue', s=suburban_city_drivers*7, edgecolors='black', label='Suburban', alpha=0.75)
rural = plt.scatter(rural_city_rides, rural_city_price, c='gold', s=rural_city_drivers*7, edgecolors='black', label='Rural', alpha=0.75)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

plt.xlim(1,42)
plt.ylim(19,45)

plt.grid(alpha=.5)

# Create a legend
plt.legend(loc='upper right')

lgnd = plt.legend(title="City Types", frameon = True, edgecolor = 'gray')
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.text(45,40,'Note: Circle size correlates with driver count per city')


# Save Figure
plt.savefig("images/kmikkbubbleplot.png")
```
