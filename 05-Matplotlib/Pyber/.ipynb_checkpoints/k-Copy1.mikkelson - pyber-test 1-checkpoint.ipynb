```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
merged = pd.merge(city_df, ride_df, on="city")

# Display the data table for preview
merged.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Split by city type
urban_type = merged.loc[merged['type']=='Urban']
suburban_type = merged.loc[merged['type']=='Suburban']
rural_type = merged.loc[merged['type']=='Rural']

#print(urban_type)

# Find x coordinates (Number of Rides per city)
# Find y coordinates (Average Fare)
# Find size (Drivers by city)


urban_cities = urban_type.groupby('city')
suburban_cities = suburban_type.groupby('city')
rural_cities = rural_type.groupby('city')


for city in urban_cities:
    urban_city_rides = urban_cities['ride_id'].count()
    urban_city_price = urban_cities['fare'].mean()
    urban_city_drivers = urban_type.set_index('city')['driver_count']

for city in suburban_cities:
    suburban_city_rides = suburban_cities['ride_id'].count()
    suburban_city_price = suburban_cities['fare'].mean()
    suburban_city_drivers = suburban_type.set_index('city')['driver_count']

for city in rural_cities:
    rural_city_rides = rural_cities['ride_id'].count()
    rural_city_price = rural_cities['fare'].mean()
    rural_city_drivers = rural_type.set_index('city')['driver_count']

#print(urban_city_rides)
#print(suburban_city_rides)


# Combine into dataframes
urban_data = ({"Urban Ride"})


# Build the scatter plots for each city types
urban = plt.scatter(urban_city_rides, urban_city_price, c='coral', s=urban_city_drivers*5, edgecolors='black', label='Urban', alpha=0.75)
suburban = plt.scatter(suburban_city_rides, suburban_city_price, c='lightblue', s=suburban_city_drivers*5, edgecolors='black', label='Suburban', alpha=0.75)
rural = plt.scatter(rural_city_rides, rural_city_price, c='gold', s=rural_city_drivers*5, edgecolors='black', label='Rural', alpha=0.75)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")


plt.xlim(0,45)
plt.ylim(18,42)

# Create a legend
plt.legend(loc='upper right')

# Incorporate a text label regarding circle size


# Save Figure
```
