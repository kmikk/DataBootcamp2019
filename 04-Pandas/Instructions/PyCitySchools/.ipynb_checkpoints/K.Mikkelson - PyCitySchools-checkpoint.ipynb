```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the needed info
total_schools = school_data_complete['school_name'].nunique()
total_students = school_data_complete['student_name'].count()
total_budget = school_data['budget'].sum()
avg_math = school_data_complete['math_score'].mean()
avg_reading = school_data_complete['reading_score'].mean()
passing_math = (school_data_complete['math_score']>=70).value_counts()[True]
math_percent = (passing_math / total_students)
passing_reading = (school_data_complete['reading_score']>=70).value_counts()[True]
reading_percent = (passing_reading/total_students)
passing_rate = (math_percent + reading_percent)/2

# Create a dataframe to hold results
district_summary = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students],
                     "Total Budget":[total_budget], "Average Math Score":[round(avg_math,2)],
                     "Average Reading Score":[round(avg_reading,2)], "% Passing Math":[math_percent],
                    "% Passing Reading":[reading_percent], "% Overall Passing Rate":[passing_rate]})

# Arrange columns
organized_summary = district_summary[["Total Schools", "Total Students", "Total Budget", "Average Math Score", 
                      "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]

# Improve formatting
organized_summary["Total Students"] = organized_summary["Total Students"].map("{:,}".format)
organized_summary["Total Budget"] = organized_summary["Total Budget"].map("${:,.2f}".format)
organized_summary["% Passing Math"] = organized_summary["% Passing Math"].map("{:.2%}".format)
organized_summary["% Passing Reading"] = organized_summary["% Passing Reading"].map("{:.2%}".format)
organized_summary["% Overall Passing Rate"] = organized_summary["% Overall Passing Rate"].map("{:.2%}".format)

# Display summary
district_summary_df = pd.DataFrame(organized_summary)
district_summary_df.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Group data by school name
by_school = school_data_complete.groupby(['school_name'])

#  * School Type
school_type = school_data.set_index('school_name')['type']

#  * Total Students
students_per_school = by_school['Student ID'].count()

#  * Total School Budget
school_budget = school_data.set_index('school_name')['budget']

#  * Per Student Budget
student_budget = school_budget / students_per_school

#  * Average Math Score
avg_math_score = by_school['math_score'].mean()

#  * Average Reading Score
avg_reading_score = by_school['reading_score'].mean()

#  * % Passing Math
#  * % Passing Reading
#  * Overall Passing Rate (Average of the above two)

passing_math_percent = []
passing_reading_percent=[]
overall_passing = []

for school, by_school_data in by_school:
    passing_math_byschool = (by_school_data['math_score']>=70).value_counts()[True]
    students_byschool = by_school_data['Student ID'].count()
    passing_math_percent.append(passing_math_byschool/students_byschool)
    passing_reading_byschool = (by_school_data['reading_score']>=70).value_counts()[True]
    passing_reading_percent.append(passing_reading_byschool/students_byschool)
    overall_passing.append((passing_math_byschool+passing_reading_byschool)/(2*students_byschool))


# Create new dataframe to hold results for referencing later in code (without formatting)
school_summary = pd.DataFrame({"School Type":school_type, "Students Per School":students_per_school, 
                               "Total Budget":school_budget, "Per Student Budget":student_budget, 
                               "Average Math Score":(round(avg_math_score,2)), "Average Reading Score":(round(avg_reading_score,2)),
                              "% Passing Math":passing_math_percent, "% Passing Reading":passing_reading_percent,
                              "% Overall Passing Rate":overall_passing})


# Create new dataframe to hold results for display
school_summary_formatted = pd.DataFrame({"School Type":school_type, "Students Per School":students_per_school, 
                               "Total Budget":school_budget, "Per Student Budget":student_budget, 
                               "Average Math Score":(round(avg_math_score,2)), "Average Reading Score":(round(avg_reading_score,2)),
                              "% Passing Math":passing_math_percent, "% Passing Reading":passing_reading_percent,
                              "% Overall Passing Rate":overall_passing})

# Improve formatting
school_summary_formatted["Students Per School"] = school_summary_formatted["Students Per School"].map("{:,}".format)
school_summary_formatted["Total Budget"] = school_summary_formatted["Total Budget"].map("${:,.2f}".format)
school_summary_formatted["Per Student Budget"] = school_summary_formatted["Per Student Budget"].map("${:,.2f}".format)
school_summary_formatted["% Passing Math"] = school_summary_formatted["% Passing Math"].map("{:.2%}".format)
school_summary_formatted["% Passing Reading"] = school_summary_formatted["% Passing Reading"].map("{:.2%}".format)
school_summary_formatted["% Overall Passing Rate"] = school_summary_formatted["% Overall Passing Rate"].map("{:.2%}".format)

```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_five = (school_summary_formatted.sort_values("% Overall Passing Rate", ascending=False)).head(5)
top_five
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_five = (school_summary_formatted.sort_values("% Overall Passing Rate", ascending=False)).tail(5)
bottom_five
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Narrow down table to just needed info
grades = school_data_complete[["school_name", "grade", "math_score", "reading_score"]]

# Rename school_name for cleaner formatting
grade_table = grades.rename(columns={"school_name":"School Name"})


# Split by grade
ninth_grade = grade_table.loc[grade_table["grade"]=='9th']
tenth_grade = grade_table.loc[grade_table["grade"]=='10th']
eleventh_grade = grade_table.loc[grade_table["grade"]=='11th']
twelvth_grade = grade_table.loc[grade_table["grade"]=='12th']

# Group scores by school
ninth_grouped = ninth_grade.groupby(["School Name"])
tenth_grouped = tenth_grade.groupby(["School Name"])
eleventh_grouped = eleventh_grade.groupby(["School Name"])
twelvth_grouped = twelvth_grade.groupby(["School Name"])

# Average the scores
ninth_avg = ninth_grouped["math_score"].mean()
tenth_avg = tenth_grouped["math_score"].mean()
eleventh_avg = eleventh_grouped["math_score"].mean()
twelvth_avg = twelvth_grouped["math_score"].mean()


# Combine into dataframe
math_by_grade = pd.DataFrame({"9th":(round(ninth_avg,2)), "10th":(round(tenth_avg)), "11th":(round(eleventh_avg,2)), "12th":(round(twelvth_avg,2))})
math_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
spending_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Narrow data frame to needed columns
by_spending = school_summary.loc[:,['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing Rate']]

# Bin and add column
by_spending['Spending by Student']=pd.cut(school_summary['Per Student Budget'], bins=spending_bins, labels=spending_names)

# Group by spending
by_spending = by_spending.groupby('Spending by Student').mean()

# Improve formatting
by_spending["Average Math Score"] = by_spending["Average Math Score"].map("{:.2f}".format)
by_spending["Average Reading Score"] = by_spending["Average Reading Score"].map("{:.2f}".format)
by_spending["% Passing Math"] = by_spending["% Passing Math"].map("{:.2%}".format)
by_spending["% Passing Reading"] = by_spending["% Passing Reading"].map("{:.2%}".format)
by_spending["% Overall Passing Rate"] = by_spending["% Overall Passing Rate"].map("{:.2%}".format)



by_spending.head()


```
